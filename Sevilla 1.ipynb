# Sevilla & All Laliga match result (2021-2022)

```python
df3=pd.read_csv("/content/all_match_results.csv",parse_dates=["Date"],index_col=0)
df3
```

```python
shome = df3[df3["HomeTeam"] == 'Sevilla FC']
saway = df3[df3["AwayTeam"] == 'Sevilla FC']
```

**All result**

**win-draw-loss home & away**

```python
def wdl_home(data):
    
    decision = []
    gf = []
    ga = []
    home = data[data["HomeTeam"] == 'Sevilla Fc']
    
    for u, v in [row.split('-') for row in shome.Result.tolist()]:
        
        u = int(u)
        v = int(v)
        if u > v:
            decision.append('W')
            gf.append(u)
            ga.append(v)
        elif u < v:
            decision.append('L')
            gf.append(u)
            ga.append(v)
        else :
            decision.append('D')
            gf.append(u)
            ga.append(v)
    home['WDL'] = decision
    home['Scored'] = gf
    home['Conceded'] = ga
    home['GD'] = home['Scored'] - home['Conceded']
    return home

def wdl_away(data):
    
    decision = []
    gf = []
    ga = []
    
    away = data[data["AwayTeam"]== 'Sevilla Fc']
    
    for u, v in [row.split('-') for row in saway.Result.tolist()]:
        
        u = int(u)
        v = int(v)
        if u > v:
            decision.append('L')
            gf.append(v)
            ga.append(u)
        elif u < v:
            decision.append('W')
            gf.append(v)
            ga.append(u)
        else:
            decision.append('D')
            gf.append(v)
            ga.append(u)
            
    away['WDL'] = decision
    away['Scored'] = gf
    away['Conceded'] = ga
    away['GD'] = away['Scored'] - away['Conceded']
    return away
```

```python
home1= wdl_home(df3)
away1= wdl_away(df3)
```

```python
fig = plt.figure(figsize=(15, 5))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
home1.WDL.value_counts().plot.bar(edgecolor='black', linewidth=1.2, label='Sevilla at home',
                                        color='red', ax=ax1);
ax1.set_ylabel('counts')
ax1.set_title('WDL home')
ax1.legend();
away1.WDL.value_counts().plot.bar(edgecolor='black', linewidth=1.2, label='Sevilla at away',
                                        color='blue', ax=ax2);
ax2.set_title('WDL away')
ax2.set_ylabel('counts')
ax2.legend();
```



*  Sevilla has won 12 matches , 6 draws,1 loose in home out of 19 matches . 
*  Sevilla has won 10 matches , 6 draws , 3 loose in away out of  19 i= matches.

**Sevilla has to improve their game in away match**
